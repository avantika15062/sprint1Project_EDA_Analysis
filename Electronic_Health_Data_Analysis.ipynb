Insight: Displaying the count of drug note type using countplot function from seaborn library.

Insight : Using patches, texts and pcts from matplotlib to design a layout for a pie chart representing drugnotetype column. So we can conclude is most patients are prescribed for medicines while admission.

```python
label_data = df['drugnotetype'].value_counts()
explode = (0.0,0.0,0.0,0.0,0.15,0.4)
plt.figure(figsize=(30,20))
patches , texts, pcts = plt.pie(label_data,
                                labels= label_data.index,
                                colors = ['blue', 'red', 'green', 'orange', 'yellow', 'purple', 'grey'],
                                pctdistance = 0.65,
                                shadow = False,
                                startangle = 0,
                                explode = explode,
                                autopct = '%1.1f%%',
                                textprops= {'fontsize': 25,
                                            'color': 'black',
                                            'weight':'bold',
                                            'family': 'serif'
                                      })
plt.setp(pcts, color = 'black')
hfont = {'fontname' : 'serif' , 'weight' : 'bold'}
plt.title ('DRUG NOTE TYPE', size = 20, **hfont)
centre_circle = plt.Circle((0,0), 0.40, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.show()
```
